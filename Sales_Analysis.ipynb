```python
import seaborn as sns
```

```python
df.drop(["Status","unnamed1"], axis=1 ,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df["Amount"]=df["Amount"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
df.rename(columns={"Marital_Status":"Shadi"} ,inplace=True)
```

```python
df.columns
```

```python
df.describe()
```

```python
#EXploratory data analysis
```

```python
#Gender
ax=sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
```

```python
sales_df=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
```

```python
sns.barplot(x='Gender' ,y='Amount',data=sales_df)
```

```python
#From Above Graph we can see the most of the buyers are feamle as compare to male 
```

```python
df.columns
```

```python
#Age
sales_age=df.groupby(['Age Group'] ,as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
```

```python
sns.barplot(x="Age Group",y="Amount",data=sales_age,)
```

```python
ax2=sns.countplot(x="Age Group",data=df ,hue="Gender")
for bar in ax2.containers:
    ax2.bar_label(bar)
```

```python
#From Above graph we can see the most buyers are 26-35 age group
```

```python
#State
sales_state=df.groupby(['State'] ,as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state=df.groupby(['State'] ,as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
#From above graph we can see the most buyers state are uttar pradesh maharashtra and karnataka #
```

```python
df.rename({"Shadi":"Marital Status"},axis=1,inplace=True)
```

```python
df.columns
```

```python
#Marital Status
ax3=sns.countplot(x="Marital Status",data=df)
sns.set(rc={'figure.figsize':(7,5)})
for bar in ax3.containers:
    ax3.bar_label(bar)
```

```python
sales_marital=df.groupby(['Marital Status','Gender'] ,as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital Status',y='Amount',data=sales_marital,hue="Gender")
```

```python
#from above graph we can see the most buyers are married(women)
```

```python
#Occupation
sns.set(rc={'figure.figsize':(20,6)})
ax4=sns.countplot(x='Occupation',data=df)
for bar in ax4.containers:
      ax4.bar_label(bar)
            
            
```

```python
sales_occupation=df.groupby(['Occupation',] ,as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_occupation)
```

```python
#from above graph we can see the most buyers from occupation are IT Sector,Healthcare,Aviation Respectively
```

```python
#Product category
sns.set(rc={'figure.figsize':(25,6)})
ax4=sns.countplot(x='Product_Category',data=df)
for bar in ax4.containers:
      ax4.bar_label(bar)
            
```

```python
sales_category=df.groupby(['Product_Category',] ,as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_category)
```

```python
#From above graph we can see the most buying category are food clothing and electronics
```

```python
sales_Id=df.groupby(['Product_ID',] ,as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Orders',data=sales_Id)
```

# the above graph we can see the Product_ID P00265242 has largest order
